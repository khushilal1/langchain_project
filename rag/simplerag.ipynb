```python
#reading the  pdf file
from langchain_community.document_loaders import PyPDFLoader
```

```python
#loading the file
loader=PyPDFLoader("attention.pdf")
```

```python
docs=loader.load()
```

```python
docs
```

```python
#convert into chunk of the pdf
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200) #objec tof thegivern class
documents=text_splitter.split_documents(docs)
```

```python
documents
```

```python
##vector enbdeeing technique and store

from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.embeddings import OpenAIEmbeddings
from langchain_community.vectorstores import Chroma

```

```python
#chroma databases
db = Chroma.from_documents(documents[:3], OllamaEmbeddings())  # first 1 document
```

```python
##vector databases
query="An attention function can be described as mapping a query"
##query to retieve data
result=db.similarity_search(query)
result[0].page_content #getting the value 
```

```python
##faiss Vector databases
from langchain_community.vectorstores import FAISS
```

```python
#faiss databases

db = FAISS.from_documents(documents[:1], OllamaEmbeddings())
##vector databases
query="An attention function can be described as mapping a query"
##query to retieve data
result=db.similarity_search(query)
result[0].page_content #getting the value 
```
